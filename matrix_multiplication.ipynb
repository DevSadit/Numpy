## Matrix Multiplication

```python
arr = np.random.rand(12)
arr2 = np.random.rand(12)
```

```python
arr.reshape(3, 4)
arr2.reshape(3, 4)
```

```python
mul = arr @ arr2
mul
```

```python
np.dot(arr, arr2)
```
