```python
arr = np.array([10, 20, 30, 40, 50])
# accesing the array with index
print(arr[0])
print(arr[-2])
```

```python
# slice the array
new_arr = arr[1:4]
print(new_arr)
```

```python
# matrix
matrix = np.array([[1, 2, 3], [11, 15, 18], [50, 100, 150]])
print(matrix.shape)

# accesing the exact number in matrix
matrix[1, 2]

# to get a whole collumn
matrix[:, 2]

# to get a whole row
matrix[1, :]
```
