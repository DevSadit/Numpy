### Boolean Filtering

```python
scores = np.array([45,82,61,90,55,72,39,99])
```

```python
scores[scores>70]
scores[scores<60]
scores[(scores>50) & (scores<80)]
```

```python
scores.mean()
scores.max()
scores.min()
scores.sum()
scores.std()
```
